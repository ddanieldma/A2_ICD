```python
import pandas as pd
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import NumeralTickFormatter
#================================================================================================
# lendo base de dados
import os
os.chdir("E:/Users/B51995/Desktop/A2_ICD")

df_cause_of_death = pd.read_csv("new_cause_of_death.csv")

output_notebook()
```

```python
df_auto_mut_ano = df_cause_of_death.select_dtypes(include = "number").groupby("Year").mean()
auto_mut_ano = ColumnDataSource(df_auto_mut_ano)

plot = figure(width = 1280, height = 720, title = "Média de mortes por auto-mutilação por ano")
plot.line(x = "Year", y = "Self-harm", source = auto_mut_ano)
plot.xaxis.formatter = NumeralTickFormatter(format = "0,0")

show(plot)
```

```python
df_auto_mut_ano = df_cause_of_death.select_dtypes(include = "number").groupby("Year").sum()
auto_mut_ano = ColumnDataSource(df_auto_mut_ano)

plot = figure(width = 1280, height = 720, title = "Total de mortes por auto-mutilação por ano")
plot.line(x = "Year", y = "Self-harm", source = auto_mut_ano)
plot.yaxis.formatter = NumeralTickFormatter(format = "0,0")

show(plot)
```

```python
# Chronic Respiratory Diseases

df_auto_mut_ano = df_cause_of_death.select_dtypes(include = "number").groupby("Year").sum()
df_mortes_china = df_cause_of_death.loc[df_cause_of_death["Country/Territory"] == "China"].groupby("Year").sum()
mortes_china = ColumnDataSource(df_mortes_china)
auto_mut_ano = ColumnDataSource(df_auto_mut_ano)

plot = figure(width = 1280, height = 720, title = "Total de mortes por doença respiratória crônica por ano")
plot.line(x = "Year", y = "Chronic Respiratory Diseases", source = auto_mut_ano, color = "Blue")
plot.line(x = "Year", y = "Chronic Respiratory Diseases", source = mortes_china, color = "Pink")
plot.yaxis.formatter = NumeralTickFormatter(format = "0,0")

show(plot)
```

```python
# Chronic Respiratory Diseases

df_auto_mut_ano = df_cause_of_death.select_dtypes(include = "number").groupby("Year").sum()
df_mortes_china = df_cause_of_death.loc[df_cause_of_death["Country/Territory"] == "China"].select_dtypes(include="number").groupby("Year").mean()
mortes_china = ColumnDataSource(df_mortes_china)
auto_mut_ano = ColumnDataSource(df_auto_mut_ano)

plot = figure(width = 1280, height = 720, title = "Média de mortes por doença respiratória crônica por país por ano")
plot.line(x = "Year", y = "Chronic Respiratory Diseases", source = auto_mut_ano, color = "Blue")
plot.line(x = "Year", y = "Chronic Respiratory Diseases", source = mortes_china, color = "Pink")
plot.yaxis.formatter = NumeralTickFormatter(format = "0,0")

show(plot)
```

```python
# mortes por violencia interpessoal para cada continente
copia_df = df_cause_of_death
color_dict = {"Africa" : "Yellow", "North America" : "Red", "South America" : "DarkRed", "Europe" : "Black", "Oceania" : "Blue", "Asia" : "Green"}
colors = []
for cada_continente in copia_df["Continente"]:
    colors.append(color_dict[cada_continente])
copia_df["color"] = colors

mortes_violencia = ColumnDataSource(copia_df)

plot = figure(width = 1280, height = 720, title = "Mortes por violencia interpessoal pelo mundo")
plot.line(x = "Year", y = "Interpersonal Violence", source = mortes_violencia)

show(plot)
```

```python
# dicionario de cores
color_dict = {"Africa" : "Yellow", "North America" : "Red", "South America" : "DarkRed", "Europe" : "Black", "Oceania" : "Blue", "Asia" : "Green"}

# agrupando por ano e continente
grouped_df = df_cause_of_death.groupby(["Continente", "Year"]).sum()

# display(grouped_df)

# colocando a cor para cada continente no dataframe
colors = []
for each_continent in grouped_df.index.values:
    colors.append(color_dict[each_continent[1]])

# adicionando coluna de cores
grouped_df["color"] = colors
# transformando em column data source
# interpersonal_violence = ColumnDataSource(grouped_df)

#selecionando nomes únicos de continente
continentes = df_cause_of_death["Continente"].unique()

plot = figure(title = "Mortes por violencia interpessoal por continente por ano", )
for continente in continentes:
    data = grouped_df.loc[continente]
    plot.circle(x = data.index.values, y = data["Interpersonal Violence"], color = data["color"], legend_label = continente)
# plot.circle(x = "Year", y = "Interpersonal Violence", color = "color", source = grouped_df)

show(plot)
```
